```python
"""
Split Data to ensure the same artist is not in the train and test sets
"""
X_train, X_test, y_train, y_test = data.iloc[:463715, 1:], data.iloc[463715:, 1:], data.iloc[:463715, 0], data.iloc[463715:, 0]

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
"""
Convert to numpy
"""
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
"""
Dimensionality reduction
plot the PCA variance ratios to see where variance plateaus
"""
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)

x = range(len(cumsum))
plt.plot(x, cumsum)
plt.show
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

pipe_lr = make_pipeline(StandardScaler(), 
                         PCA(n_components=0.95),
                         LinearRegression())

pipe_lr.fit(X_train, y_train)
pred = pipe_lr.predict(X_train)

right = 0
for i in range(len(pred)):
    if  abs(pred[i] - y_train[i]) <= 5:
        right+=1
score = right / len(pred)
print(f'Accuracy score: {score}')
```

```python
"""
First try linear regression
"""
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train, y_train)
prediction = lr.predict(X_train)
prediction = prediction.astype(int)

right = 0
for i in range(len(prediction)):
    if prediction[i] == y_train[i]:
        right+=1
score = right / len(prediction)
print(f'Accuracy score: {score}')
```

```python
"""
Prediction was very low for the exact year.
Lets see if it can predict the correct decade range
"""
right = 0
for i in range(len(prediction)):
    if abs(prediction[i] - y_train[i]) <= 5:
        right+=1
score = right / len(prediction)
print(f'Accuracy score: {score}')
```

```python
from sklearn.linear_model import SGDRegressor

pipe_SGD = make_pipeline(StandardScaler(), 
                         PCA(n_components=0.95),
                         SGDRegressor(max_iter=1000, tol=1e-3, penalty=None))

pipe_SGD.fit(X_train, y_train)
pred = pipe_SGD.predict(X_train)

right = 0
for i in range(len(pred)):
    if  abs(pred[i] - y_train[i]) <= 5:
        right+=1
score = right / len(pred)
print(f'Accuracy score: {score}')
```

```python
from sklearn.linear_model import Ridge

pipe_ridge = make_pipeline(StandardScaler(), 
                         PCA(n_components=0.95),
                         Ridge(alpha=0.1))

pipe_ridge.fit(X_train, y_train)
pred = pipe_ridge.predict(X_train)

right = 0
for i in range(len(pred)):
    if  abs(pred[i] - y_train[i]) <= 5:
        right+=1
score = right / len(pred)
print(f'Accuracy score: {score}')
```
